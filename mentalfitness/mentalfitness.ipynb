## Importing Libraries 

```python
# Import necessary libraries
import pandas as pd             # For data manipulation and analysis using dataframes
import numpy as np              # For numerical computing and array operations
from sklearn.preprocessing import LabelEncoder   # For encoding categorical variables
import tensorflow as tf         # For building and training machine learning models
from tensorflow.keras import Sequential         # For creating a sequential model in Keras
from tensorflow.keras.layers import Dense       # For adding dense layers to the model
import matplotlib.pyplot as plt  # For data visualization using plots
```

## Loading Data

```python


# Read the first CSV file into a pandas DataFrame
data_1 = pd.read_csv('mental-and-substance-use-as-share-of-disease -AI.csv')

# The 'data_1' DataFrame now contains information related to the share of mental and substance use disorders in disease prevalence

# Read the second CSV file into another pandas DataFrame
data_2 = pd.read_csv('prevalence-by-mental-and-substance-use-disorder _AI.csv')

# The 'data_2' DataFrame now contains information related to the prevalence of mental and substance use disorders
```

```python
# Merge the two DataFrames 'data_1' and 'data_2' based on their common columns
data = pd.merge(data_1, data_2)

# The resulting 'data' DataFrame contains the merged data from 'data_1' and 'data_2', with common columns aligned.

# Depending on the data in 'data_1' and 'data_2', the merge may be based on one or more common columns.

# By default, pandas uses inner join to merge the DataFrames, which keeps only the rows with matching keys in both DataFrames.

# The 'data' DataFrame now contains the combined information from both 'data_1' and 'data_2'.
data
```

```python
# Check for missing values in the 'data' DataFrame
data.isnull().sum()

# The 'isnull()' function returns a DataFrame with the same shape as 'data', where each entry is a boolean value
# indicating whether the corresponding element is missing (True) or not missing (False).

# The 'sum()' function then calculates the total number of missing values for each column in the DataFrame.

# The resulting output displays the count of missing values for each column in the 'data' DataFrame.
data
```

```python
# Remove the 'Code' column from the 'data' DataFrame
data.drop('Code', axis=1, inplace=True)

# The 'drop()' function is used to remove columns from the DataFrame.
# The first argument ('Code') specifies the name of the column to be dropped.
# The second argument ('axis=1') indicates that we want to drop a column (axis=0 is for dropping rows).
# The third argument ('inplace=True') modifies the 'data' DataFrame in-place without returning a new DataFrame.

# The 'Code' column has been removed from the 'data' DataFrame.

# Remove the 'Year' column from the 'data' DataFrame
data.drop('Year', axis=1, inplace=True)

# Similarly, the 'Year' column is removed from the 'data' DataFrame.
# After executing this, the 'data' DataFrame will no longer contain the 'Year' column.

data
```

```python
# Extract the column 'DALYs (Disability-Adjusted Life Years) - Mental disorders - Sex: Both - Age: All Ages (Percent)' from the 'data' DataFrame
column = data.pop('DALYs (Disability-Adjusted Life Years) - Mental disorders - Sex: Both - Age: All Ages (Percent)')

# The 'pop()' function is used to remove a column from the DataFrame and return it.
# The argument ('DALYs (Disability-Adjusted Life Years) - Mental disorders - Sex: Both - Age: All Ages (Percent)') specifies the name of the column to be extracted.

# The extracted column is now stored in the variable 'column'.

# Add the extracted column back to the 'data' DataFrame with the same name
data['DALYs (Disability-Adjusted Life Years) - Mental disorders - Sex: Both - Age: All Ages (Percent)'] = column

# The column stored in the variable 'column' is added back to the 'data' DataFrame with the same column name.
# After executing this, the 'data' DataFrame will have the column 'DALYs (Disability-Adjusted Life Years) - Mental disorders - Sex: Both - Age: All Ages (Percent)' twice,
# once with the original values and the other with the values obtained from the variable 'column'.
```

```python
# Rename the columns of the 'data' DataFrame with more meaningful names
data = data.set_axis(['Country', 'Schizophrenia', 'Bipolar', 'Eating', 'Anxiety', 'Drug', 'Depressive', 'Alcohol', 'mental_fitness'], axis=1, copy=False)

# The 'set_axis()' function is used to rename the columns of the DataFrame.
# The first argument specifies the new column names as a list in the order they should be assigned.
# In this case, the columns are renamed as follows:
# - 'Country': The column containing country names
# - 'Schizophrenia': The column representing data related to schizophrenia
# - 'Bipolar': The column representing data related to bipolar disorder
# - 'Eating': The column representing data related to eating disorders
# - 'Anxiety': The column representing data related to anxiety disorders
# - 'Drug': The column representing data related to drug use disorders
# - 'Depressive': The column representing data related to depressive disorders
# - 'Alcohol': The column representing data related to alcohol use disorders
# - 'mental_fitness': The column representing data related to mental fitness

# The columns of the 'data' DataFrame have been renamed according to the provided list of names.
# By using 'copy=False', the operation is performed in-place, and no new DataFrame is created.
```

## Changing all labels to Numric Labels

```python
# Create a LabelEncoder object
l = LabelEncoder()

# The LabelEncoder is used to convert categorical values in a column to numerical representations.

# Iterate through each column in the 'data' DataFrame
for i in data.columns:
    # Check if the data type of the column is 'object', indicating that it contains categorical values
    if data[i].dtypes == 'object':
        # Use the LabelEncoder to transform the categorical values in the column to numerical representation
        data[i] = l.fit_transform(data[i])

        # The 'fit_transform()' method of the LabelEncoder first fits the encoder to the unique categorical values in the column,
        # assigning each unique value a unique numerical label.
        # Then, it transforms the original categorical values in the column to their corresponding numerical labels.

# After this loop, all columns containing categorical values in the 'data' DataFrame have been converted to numerical representations
# using the LabelEncoder. This makes the data suitable for training machine learning models that require numerical inputs.
```

## Splitting Data into Train & Dev Sets

```python
# Create a shuffled version of the 'data' DataFrame
Shuffledata = data.sample(frac=1, random_state=41)

# The 'sample()' function is used to randomly shuffle the rows of the DataFrame.

# The 'frac=1' argument specifies that we want to sample the entire DataFrame, which effectively shuffles all rows.
# Setting 'frac=1' means we want to include all rows in the shuffled dataset.

# The 'random_state=41' argument sets a random seed to ensure reproducibility. If the same seed is used in future runs,
# the shuffling will produce the same order of rows.

# The resulting 'Shuffle_data' DataFrame contains the original 'data' DataFrame with its rows randomly shuffled.
# This shuffling is useful for tasks such as creating training and testing datasets or ensuring randomness in data processing.
```

```python
Shuffledata
```

```python
# Extract the input features (X) for training, development, and testing datasets from the shuffled 'Shuffledata' DataFrame.

# Training data (X_train):
# Select the first 5472 rows and the first 8 columns (0 to 7) from the 'Shuffledata' DataFrame.
X_train = np.array(Shuffledata.iloc[:5472, 0:8])

# Development data (X_dev):
# Select rows from index 5472 to 6155 (5472:6156) and the first 8 columns (0 to 7) from the 'Shuffledata' DataFrame.
X_dev = np.array(Shuffledata.iloc[5472:6156, 0:8])

# Testing data (X_test):
# Select rows from index 6156 onwards and the first 8 columns (0 to 7) from the 'Shuffledata' DataFrame.
X_test = np.array(Shuffledata.iloc[6156:, 0:8])

# The 'iloc' function is used to select specific rows and columns based on their index positions.
# The first argument of 'iloc' specifies the rows to be selected, and the second argument specifies the columns.

# 'np.array()' converts the selected data into NumPy arrays, which are commonly used in machine learning models.

# The shapes of the extracted datasets are then printed to show the number of samples and features in each dataset.

print("X_train : ", X_train.shape)  # Output the shape of the training dataset (number of samples, number of features)
print("X_dev : ", X_dev.shape)      # Output the shape of the development dataset (number of samples, number of features)
print("X_test : ", X_test.shape)    # Output the shape of the testing dataset (number of samples, number of features)
```

```python
# Extract the target variable (Y) for training, development, and testing datasets from the shuffled 'Shuffledata' DataFrame.

# Training data (Y_train):
# Select the values from the column at index 8 (column with index 8) from the 'Shuffledata' DataFrame for training target variable.
# The '.reshape(-1,1)' is used to convert the 1-dimensional array to a 2-dimensional column vector, which is required for some machine learning models.
Y_train = np.array(Shuffledata.iloc[:5472, 8]).reshape(-1, 1)

# Development data (Y_dev):
# Select values from index 5472 to 6155 (5472:6156) in the column at index 8 from the 'Shuffledata' DataFrame for development target variable.
# The '.reshape(-1,1)' converts the 1-dimensional array to a 2-dimensional column vector.
Y_dev = np.array(Shuffledata.iloc[5472:6156, 8]).reshape(-1, 1)

# Testing data (Y_test):
# Select values from index 6156 onwards in the column at index 8 from the 'Shuffledata' DataFrame for testing target variable.
# The '.reshape(-1,1)' converts the 1-dimensional array to a 2-dimensional column vector.
Y_test = np.array(Shuffledata.iloc[6156:, 8]).reshape(-1, 1)

# The '.reshape(-1,1)' is used here to ensure that the target variable arrays have the appropriate shape for training machine learning models.

# The shapes of the extracted target variable arrays are then printed to show the number of samples in each dataset.

print("Y_train : ", Y_train.shape)  # Output the shape of the training target variable array (number of samples, 1)
print("Y_dev : ", Y_dev.shape)      # Output the shape of the development target variable array (number of samples, 1)
print("Y_test : ", Y_test.shape)    # Output the shape of the testing target variable array (number of samples, 1)
```

## Model 

```python
# Create a sequential neural network model using Keras

# The 'Sequential' model in Keras is a linear stack of layers where each layer has exactly one input tensor and one output tensor.

# The model consists of two dense (fully connected) layers:

# 1. Dense layer with 100 units and sigmoid activation function:
# - 'units=100': The layer will have 100 neurons or units.
# - 'activation='sigmoid'': The activation function used for the neurons in this layer is the sigmoid function.
# - 'input_shape=(8,)': The input shape specifies the number of features (8 in this case) for the input data.
#   This is the first layer, so it needs to know the shape of the input data.

# 2. Dense layer with 1 unit and linear activation function:
# - 'units=1': The output layer will have 1 neuron or unit, as it's a regression task with a single output value.
# - 'activation='linear'': The activation function used for the neuron in this layer is the linear function (identity function).
#   In regression tasks, we use the linear activation to directly predict a continuous value.

# Note: The activation functions and the number of units in the dense layers can be modified based on the requirements of the specific task.

# The 'model' variable holds the sequential neural network model defined above.

model = Sequential([
    Dense(units=100, activation='relu', input_shape=(8,)),
    Dense(units=1, activation='linear')
])
```

### Architecture

```python
model.summary()
```

### Compiling Model

```python
# Compile the sequential neural network model

# The 'compile()' function in Keras is used to configure the learning process for the model.

# Loss function:
# 'loss='mean_squared_error'' specifies the loss function to measure the difference between the predicted output and the actual target.
# For regression tasks, mean squared error (MSE) is a common loss function used to measure the average squared difference between predictions and true values.

# Optimizer:
# 'optimizer='adam'' specifies the optimization algorithm used to update the weights of the model during training.
# Adam (Adaptive Moment Estimation) is a popular optimization algorithm that combines the advantages of both AdaGrad and RMSProp.

# The model is now configured and ready for training with the specified loss function and optimizer.

model.compile(loss='mean_squared_error', optimizer='adam')
```

### Train Model

```python
# Train the sequential neural network model on the training data

# The 'fit()' function in Keras is used to train the model on the provided training data.

# Training data:
# 'x=X_train': The input features (X) for the training data.
# 'y=Y_train': The target variable (Y) for the training data.

# Validation data:
# 'validation_data=(X_dev, Y_dev)': The validation data used to evaluate the model after each epoch during training.
# X_dev contains the input features for the development dataset, and Y_dev contains the corresponding target values.

# Epochs:
# 'epochs=40': The number of times the model is trained on the entire training dataset.
# One epoch is a complete iteration through the entire dataset.

# The 'fit()' function trains the model using the training data and validates its performance using the validation data.
# It returns a 'Train' object that contains information about the training process, such as loss and accuracy for each epoch.

# The model is now trained and ready for evaluation and prediction tasks.

Train = model.fit(x=X_train, y=Y_train, validation_data=(X_dev, Y_dev), epochs=40)
```

```python
# Plot the training and validation loss over epochs

# 'Train' object contains historical training information, including loss values for each epoch.

# Extract training and validation loss values from the 'Train' object.
loss = Train.history['loss']        # Training loss values for each epoch
val_loss = Train.history['val_loss']   # Validation loss values for each epoch

# 'epochs' is a range of integers from 1 to the number of epochs.
epochs = range(1, len(loss) + 1)

# Create a line plot to visualize the training and validation loss over epochs.
# 'plt.plot()' is used to draw the lines on the plot.
# 'b' stands for blue color and 'r' stands for red color.

plt.plot(epochs, loss, 'b', label='Training Loss')     # Plot the training loss in blue color
plt.plot(epochs, val_loss, 'r', label='Validation Loss')  # Plot the validation loss in red color

# Set plot title and labels for the x-axis and y-axis.
plt.title('Epochs vs. Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')

# Display a legend to differentiate between training and validation loss lines.
plt.legend()

# Display the plot with the training and validation loss over epochs.
plt.show()
```

## Evaluation

```python
# Evaluate model performance on the test dataset

# The 'evaluate()' function in Keras is used to evaluate the trained model on new, unseen data, such as the test dataset.

# Test data:
# 'X_test': The input features for the test dataset.
# 'Y_test': The target variable for the test dataset.

# The 'evaluate()' function calculates the loss (error) of the model's predictions on the test dataset.

# The computed loss value is assigned to the variable 'loss'.

loss = model.evaluate(X_test, Y_test)

# The 'loss' variable now holds the evaluation result (loss value) on the test dataset.

# Print the evaluation results
print("Test Loss:", loss)

# The 'print()' function is used to display the test loss value, which indicates the model's performance on the unseen test data.
# The lower the loss value, the better the model's performance on the test dataset.
```

```python
# Make predictions on the test dataset and print the predicted and actual values side by side

# 'model.predict()' is used to make predictions using the trained model on the test dataset.

# Predicted values:
# 'y_pred': An array containing the predicted values for the test dataset obtained from the model.

y_pred = model.predict(X_test)

# Iterate through each sample in the test dataset using a loop
for i in range(len(X_test)):
    # Print the predicted value and actual value side by side for each sample
    # 'y_pred[i]' represents the predicted value for the i-th sample in the test dataset
    # 'Y_test[i]' represents the actual target value (ground truth) for the i-th sample in the test dataset

    print("Predicted:", y_pred[i], "\tActual:", Y_test[i])

# The loop iterates through all samples in the test dataset and displays the predicted and actual values for each sample side by side.
# This allows you to visually compare the model's predictions with the true target values.
```
